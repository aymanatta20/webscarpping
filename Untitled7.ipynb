```python
import requests
```

```python
from bs4 import BeautifulSoup
```

```python
url = "https://www.yallakora.com/match-center/%D9%85%D8%B1%D9%83%D8%B2-%D8%A7%D9%84%D9%85%D8%A8%D8%A7%D8%B1%D9%8A%D8%A7%D8%AA?date=9/21/2024#"
```

```python
requests.get(url)
```

```python
website = requests.get(url).text
```

```python
website
```

```python
mywebsite = BeautifulSoup(website , "html")
mywebsite
```

```python
print(mywebsite.prettify())
```

```python
matchs = mywebsite.find_all("div" , class_ = "matchCard")
matchs
```

```python
len(matchs)
```

```python
matchlist = matchs[0].find("h2").text.strip()
matchlist
```

```python
for i in range(len(matchs)): 
    print(matchs[i].find("h2").text.strip())
```

```python
matchtitle = matchs[0].find_all("div" , class_="teamsData")
matchtitle
```

```python
len(matchtitle)
```

```python
for i in range(len(matchs)) : 
    teama = matchs[i].find("div",class_="teams teamA").text.strip()
    teamb = matchs[i].find("div" , class_="teams teamB").text.strip() 
    results = matchs[i].find("div" , class_="MResult") 
    scoreA = results.find_all("span" , class_="score")[0].text.strip() 
    scoreB = results.find_all("span" , class_="score")[1].text.strip() 
    time = results.find("span" , {"class":"time"}).text.strip()
    print(f" {teama} {scoreA}  -  {scoreB} {teamb}\n\t {[time]}")
    
 
```

```python
for i in range(len(matchs)): 
    print(matchs[i].find("h2").text.strip())
    list = matchs[i].find_all("div" , class_="teamsData")
    for j in range(len(list)) : 
        teamA = list[j].find("div" , class_="teams teamA").text.strip() 
        teamB = list[j].find("div" , class_="teams teamB").text.strip() 
        results = list[j].find("div" , class_="MResult" ).find_all("span" , class_= "score") 
        score = f"{results[0].text.strip()} - {results[1].text.strip()}"
        time = list[j].find("div" , class_= "MResult").find("span" , class_= "time").text.strip() 
        print(f"{teamA} {score} {teamB} \n\t {time}")
    print("=="*20) 

       
        
```
