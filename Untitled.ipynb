```python
import requests
```

```python
from bs4 import BeautifulSoup
```

```python
url = "https://en.wikipedia.org/wiki/List_of_largest_companies_in_the_United_States_by_revenue"
```

```python
requests.get(url)
```

```python
website = requests.get(url).text
website
```

```python
mywebsite = BeautifulSoup(website , "html")
mywebsite
```

```python
print(mywebsite.prettify())
```

```python
table = mywebsite.find_all("table" , class_="wikitable")[0]
print(table.prettify())
```

```python
companies = []

for row in table.find('tbody').find_all('tr'):
    columns = row.find_all('td')
    if len(columns) >= 2:
        company_name = columns[1].text.strip()
        revenue = columns[3].get_text(strip=True)
        employees = columns[5].text.strip()

        companies.append({'Company': company_name, 'Revenue': revenue, 'Employees':employees})  
```

```python
companies
```

```python
for company in companies:
    print(f"Company: {company['Company']}")
    print(f"Revenue: {company['Revenue']}")
    print(f"Employees: {company['Employees']}")
    print("----------")
```

```python
import csv 
csvfile = open("wikipedia.csv" , "w" , encoding = "utf-8") 
writer = csv.writer(csvfile)
writer.writerow(["company_name , Revenue , employees"])
```
