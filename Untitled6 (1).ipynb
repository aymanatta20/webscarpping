```python
jobs = mywebsite.find_all("div",class_="css-1gatmva e1v1l3u10")
len(jobs)
```

```python
for job in jobs : 
    jobtitle = job.find("h2" , class_="css-m604qf").a.text
    company = job.find("a" , class_="css-17s97q8").text
    jobtype = job.find("span", class_="css-1ve4b75 eoyjyou0").text
    companylocation = job.find("span" , class_="css-5wys0k").text
    print(jobtitle,company,companylocation,jobtype)
    
```

```python
for i in range (8) : 
    url = f"https://wuzzuf.net/search/jobs/?a=hpb&q=web%20development&start={i}"
    website = requests.get(url).text
    mywebsite = BeautifulSoup(website,"html") 
    jobs = mywebsite.find_all("div",class_="css-1gatmva e1v1l3u10")
    for job in jobs : 
        jobtitle = job.find("h2" , class_="css-m604qf").a.text
        company = job.find("a" , class_="css-17s97q8").text
        jobtype = job.find("span", class_="css-1ve4b75 eoyjyou0").text
        companylocation = job.find("span" , class_="css-5wys0k").text
        print(jobtitle,company,companylocation,jobtype)
        
```

```python
import csv 
csvfile = open("wuzzuf.cvs" , "w" , encoding = "utf-8")
writer = csv.writer(csvfile)
writer.writerow(["jobtitle, company,companylocation,jobtype"])
```
